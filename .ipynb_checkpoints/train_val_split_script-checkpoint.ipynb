```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Dont use this/bank_data.csv", index_col=0)
data
```

```python
X = data.drop(["y"], axis=1)
y = data["y"]
```

```python
X
```

```python
y
```

```python
y.unique()
```

```python
X_train, X_val, y_train, y_val = train_test_split(
        X, y, stratify=y, test_size=0.3, random_state = 42)
y_train.map({"yes": 1, "no": 0}).hist()
```

```python
y_val.map({"yes": 1, "no": 0}).hist()
```

```python
data_train = X_train
data_train["y"] = y_train
data_train.reset_index(drop = True, inplace = True)
data_train
```

```python
data_val = X_val
data_val["y"] = y_val
data_val.reset_index(drop = True, inplace = True)
data_val
```

```python
data_train.to_csv("bank_train_data.csv")
data_val.to_csv("bank_val_data.csv")
```
