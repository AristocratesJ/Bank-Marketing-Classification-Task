```python
# Funkcja pomocnicza do pokazywania procentÃ³w
def without_hue(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.2
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
```

```python
# Binarne
fig, axs = plt.subplots(2, 2, figsize = (10, 10))
axs = axs.flatten()
fig.suptitle("Zmienne binarne")

for i, col in enumerate(binary):
    sns.countplot(data = bank_data, x = col, ax=axs[i], color = "brown", order=["no", "yes"])
    without_hue(axs[i], bank_data[col])
```

- Jednak tego typu ploty majÄ… problem z tym, Å¼e najwaÅ¼niejsze okazujÄ… siÄ™ dane na koÅ„cu "wyplotowane", czyli zobaczymy gÅ‚Ã³wnie obserwacje z koÅ„ca naszej ramki danych(gdyÅ¼ program robiÄ…cy wykres nakÅ‚ada na siebie punkty). Å»eby mieÄ‡ pewnoÅ›Ä‡ co do oceny, czy moÅ¼na rozdzielaÄ‡ wg danej grupy przydaÅ‚oby siÄ™ zliczaÄ‡ zatem jakÄ…Å› proporcjÄ™ w danych obszarach. StÄ…d pomysÅ‚ na kolejne heatmapy

```python
bins = 10
fig, axs = plt.subplots(7, 7, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)

for i, col1 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
    for j, col2 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
        if i != j:
            df = pd.DataFrame({col1: pd.cut(bank_data[col1], bins = bins),\
                  col2: pd.cut(bank_data[col2], bins = bins),\
                  "y": bank_data["y"].copy()})
            df = df.groupby([col1, col2]).mean().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            fig.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
```

- Tutaj zastosowane byÅ‚y rÃ³wny przedziaÅ‚ kubeÅ‚kÃ³w co spowodowaÅ‚o nadmiernie skupienie siÄ™ nasycenia kolorÃ³w (a zatem teÅ¼ naszych oczu) na outlierach, co nie powinno byÄ‡ celem tego wykresu. StÄ…d zrobimy rÃ³wne kubeÅ‚ki i zobaczymy co siÄ™ stanie

```python
#Tutaj dla rÃ³wnych binÃ³w (wg ilosci wpadajacych do binow)
bins = 10
fig, axs = plt.subplots(7, 7, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)
fig.suptitle("Heatmapy dla zmiennych numerycznych")

for i, col1 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
    for j, col2 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
        if i != j:
            df = pd.DataFrame({col1: pd.qcut(bank_data[col1], q = bins, duplicates = "drop"),\
                  col2: pd.qcut(bank_data[col2], q = bins, duplicates = "drop"),\
                  "y": bank_data["y"].copy()})
            df = df.groupby([col1, col2]).mean().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            fig.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
```

- Po wrzuceniu outlierÃ³w do skrajnych kubeÅ‚kÃ³w obraz wydaje siÄ™ byÄ‡ bardziej rzeczywisty.
- WidaÄ‡ ewidentnÄ… zaleÅ¼noÅ›Ä‡ dla cechy duration, Å¼e ma ona wysoki wpÅ‚yw prawdopodobieÅ„stwo zaÅ‚oÅ¼enia depozytu nie zaleÅ¼nie od drugiej cechy
- W niektÃ³rych subplotach "zmniejszyÅ‚a siÄ™" liczba kubeÅ‚kÃ³w na niektÃ³rych osiach. Ma to zwiÄ…zek z tym, Å¼e funkcja qcut stara siÄ™ podzieliÄ‡ kubeÅ‚ki na rÃ³wne. Jednak na poczÄ…tku zauwaÅ¼yliÅ›my, Å¼e wiele z naszych cech ma mocno skoÅ›ny rozkÅ‚ad (w dodatku sÄ… to inty) stÄ…d nie mÃ³gÅ‚ rÃ³wno podzieliÄ‡ na wczeÅ›niej zaznaczonÄ… iloÅ›Ä‡ unikalnych kubeÅ‚kÃ³w. StÄ…d daÅ‚em moÅ¼liwoÅ›Ä‡ zredukowania iloÅ›ci kubeÅ‚kÃ³w w takim przypadku (duplicates="drop"), aby obsÅ‚uÅ¼yÄ‡ taki przypadek
- W takiej reprezentacji moÅ¼na zauwaÅ¼yÄ‡ teÅ¼, Å¼e previous mocno wpÅ‚ywa na prawdopodobieÅ„stwo przy uwzglÄ™dnieniu duration. OczywiÅ›cie moÅ¼na to wytÅ‚umaczyÄ‡ tym, Å¼e czÅ‚owiek, ktÃ³ry miaÅ‚ juÅ¼ stycznoÅ›Ä‡ poprzednio z tym bankiem chÄ™tniej dowie siÄ™ wiÄ™cej o ofercie i koÅ„cowo zdecyduje siÄ™ na proponowany produkt/zaÅ‚oÅ¼enie depozytu.
- OprÃ³cz tego okazuje siÄ™, Å¼e zmienna balance ma wpÅ‚yw tzn, Å¼e ludzie, ktÃ³rzy wychodzÄ… na duÅ¼ym plusie r/r chÄ™tniej zaÅ‚oÅ¼Ä… konto jeÅ›li wczeÅ›niej mieli kontakt z danym bankiem (plot previous/balance)

```python
fig, axs = plt.subplots(9, 9, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)
fig.suptitle("Heatmapy dla zmiennych kategorycznych")

for i, col1 in enumerate(["default", "housing", "loan", "education", "month", "job", "marital", "contact", "poutcome"]):
    for j, col2 in enumerate(["default", "housing", "loan", "education", "month", "job", "marital", "contact", "poutcome"]):
        if i != j:
            df = bank_data[[col1, col2, "y"]].groupby([col1, col2]).mean().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            if i == 0 and j == 8:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            elif i == 0 and j != 8:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, xticklabels=False)
            elif j == 8 and i != 0:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, yticklabels=False)
            else:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, xticklabels=False, yticklabels = False)
```

- te ploty sÄ… dla zmiennych kategorycznych, czyli kubeÅ‚ki nie sÄ… podzielone rÃ³wno, stÄ…d nasycenie kolorÃ³w jest wysokie
- widaÄ‡ zaleÅ¼noÅ›ci ewidentne od poutcome w kaÅ¼dym przypadku w przypadku sukcesu
- ploty te zostaÅ‚y zrobione, Å¼eby ewentualnie zajrzeÄ‡ do Å›rodka do poszczegÃ³lnych plotÃ³w pojedynczo

```python
# TÄ… funkcjÄ… moÅ¼emy wchodziÄ‡ do poszegÃ³lnych plotÃ³w tak, Å¼eby zobaczyÄ‡ dla jakich kategorii co jest
def make_categorical_heatmap(col1, col2):
    df = bank_data[[col1, col2, "y"]].groupby([col1, col2]).mean().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
    annot = bank_data[[col1, col2, "y"]].groupby([col1, col2]).count().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
    fig = sns.heatmap(df, cmap = sns.cm.rocket_r, vmin=0, vmax=1, annot = annot, fmt='g')
    fig.set_xticklabels(fig.get_xticklabels(), rotation=45, horizontalalignment='right')
    plt.show()
```

```python
# Liczba w Å›rodku informuje nas ile w tym kubeÅ‚ku razem byÅ‚o obserwacji (bo wiemy, Å¼e kategorie nie sÄ… rÃ³wno podzielone)
# W ten sposÃ³b kolory nas nie zmylÄ…, jeÅ›li bÄ™dÄ… to outliery (rzadkie kombinacje)
make_categorical_heatmap("job", "marital")
```

```python
# WidaÄ‡, Å¼e ludzie niepracujÄ…cy (retired i student) nie majÄ…cy kredytu hipotecznego majÄ… sporÄ… szanse na zaÅ‚oÅ¼enie konta
make_categorical_heatmap("job", "housing")
```

```python
make_categorical_heatmap("month", "education")
```
