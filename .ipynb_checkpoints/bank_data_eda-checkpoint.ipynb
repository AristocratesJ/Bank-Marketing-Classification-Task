```python
# Funkcja pomocnicza do pokazywania procentów
def without_hue(ax, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.2
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
```

```python
# Binarne
fig, axs = plt.subplots(2, 2, figsize = (10, 10))
axs = axs.flatten()
fig.suptitle("Zmienne binarne")

for i, col in enumerate(binary):
    sns.countplot(data = bank_data, x = col, ax=axs[i], color = "brown", order=["no", "yes"])
    without_hue(axs[i], bank_data[col])
```

- Możliwe, że połączenie zmiennych duration z campaign spowoduje łatwiejsze rozdzielenie ludzi na grupy
- Tak samo możliwe z age
- Jednak tego typu ploty mają problem z tym, że najważniejsze okazują się dane na końcu "wyplotowane", czyli zobaczymy głównie obserwacje z końca naszej ramki danych(gdyż program robiący wykres nakłada na siebie punkty). Żeby mieć pewność co do oceny, czy można rozdzielać wg danej grupy przydałoby się zliczać zatem jakąś proporcję w danych obszarach. Stąd pomysł na kolejne heatmapy

```python
bins = 10
fig, axs = plt.subplots(7, 7, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)

for i, col1 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
    for j, col2 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
        if i != j:
            df = pd.DataFrame({col1: pd.cut(bank_data[col1], bins = bins),\
                  col2: pd.cut(bank_data[col2], bins = bins),\
                  "y": bank_data["y"].copy()})
            df = df.groupby([col1, col2]).mean().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            fig.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
```

- Tutaj zastosowane były równy przedział kubełków co spowodowało nadmiernie skupienie się nasycenia kolorów (a zatem też naszych oczu) na outlierach, co nie powinno być celem tego wykresu. Stąd zrobimy równe kubełki i zobaczymy co się stanie

```python
#Tutaj dla równych binów (wg ilosci wpadajacych do binow)
bins = 10
fig, axs = plt.subplots(7, 7, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)
fig.suptitle("Heatmapy dla zmiennych numerycznych")

for i, col1 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
    for j, col2 in enumerate(["age", "balance", "day", "duration", "campaign", "pdays", "previous"]):
        if i != j:
            df = pd.DataFrame({col1: pd.qcut(bank_data[col1], q = bins, duplicates = "drop"),\
                  col2: pd.qcut(bank_data[col2], q = bins, duplicates = "drop"),\
                  "y": bank_data["y"].copy()})
            df = df.groupby([col1, col2]).mean().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            fig.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
```

- Po wrzuceniu outlierów do skrajnych kubełków obraz wydaje się być bardziej rzeczywisty.
- Widać ewidentną zależność dla cechy duration, że ma ona wysoki wpływ prawdopodobieństwo założenia depozytu nie zależnie od drugiej cechy
- W niektórych subplotach "zmniejszyła się" liczba kubełków na niektórych osiach. Ma to związek z tym, że funkcja qcut stara się podzielić kubełki na równe. Jednak na początku zauważyliśmy, że wiele z naszych cech ma mocno skośny rozkład (w dodatku są to inty) stąd nie mógł równo podzielić na wcześniej zaznaczoną ilość unikalnych kubełków. Stąd dałem możliwość zredukowania ilości kubełków w takim przypadku (duplicates="drop"), aby obsłużyć taki przypadek
- W takiej reprezentacji można zauważyć też, że previous mocno wpływa na prawdopodobieństwo przy uwzględnieniu duration. Oczywiście można to wytłumaczyć tym, że człowiek, który miał już styczność poprzednio z tym bankiem chętniej dowie się więcej o ofercie i końcowo zdecyduje się na proponowany produkt/założenie depozytu.
- Oprócz tego okazuje się, że zmienna balance ma wpływ tzn, że ludzie, którzy wychodzą na dużym plusie r/r chętniej założą konto jeśli wcześniej mieli kontakt z danym bankiem (plot previous/balance)

```python
fig, axs = plt.subplots(9, 9, figsize = (20, 20))
fig.subplots_adjust(hspace = 0.4)
fig.suptitle("Heatmapy dla zmiennych kategorycznych")

for i, col1 in enumerate(["default", "housing", "loan", "education", "month", "job", "marital", "contact", "poutcome"]):
    for j, col2 in enumerate(["default", "housing", "loan", "education", "month", "job", "marital", "contact", "poutcome"]):
        if i != j:
            df = bank_data[[col1, col2, "y"]].groupby([col1, col2]).mean().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
            if i == 0 and j == 8:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1)
            elif i == 0 and j != 8:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, xticklabels=False)
            elif j == 8 and i != 0:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, yticklabels=False)
            else:
                fig = sns.heatmap(df, cmap = sns.cm.rocket_r, ax = axs[i][j], vmin=0, vmax=1, xticklabels=False, yticklabels = False)
```

- te ploty są dla zmiennych kategorycznych, czyli kubełki nie są podzielone równo, stąd nasycenie kolorów jest wysokie
- widać zależności ewidentne od poutcome w każdym przypadku w przypadku sukcesu
- ploty te zostały zrobione, żeby ewentualnie zajrzeć do środka do poszczególnych plotów pojedynczo

```python
# Tą funkcją możemy wchodzić do poszególnych plotów tak, żeby zobaczyć dla jakich kategorii co jest
def make_categorical_heatmap(col1, col2):
    df = bank_data[[col1, col2, "y"]].groupby([col1, col2]).mean().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
    annot = bank_data[[col1, col2, "y"]].groupby([col1, col2]).count().unstack(fill_value=0).stack().reset_index().pivot(index=col1, columns=col2, values="y").sort_index(ascending = False)
    fig = sns.heatmap(df, cmap = sns.cm.rocket_r, vmin=0, vmax=1, annot = annot, fmt='g')
    fig.set_xticklabels(fig.get_xticklabels(), rotation=45, horizontalalignment='right')
    plt.show()
```

```python
# Liczba w środku informuje nas ile w tym kubełku razem było obserwacji (bo wiemy, że kategorie nie są równo podzielone)
# W ten sposób kolory nas nie zmylą, jeśli będą to outliery (rzadkie kombinacje)
make_categorical_heatmap("job", "marital")
```

```python
# Widać, że ludzie niepracujący (retired i student) nie mający kredytu hipotecznego mają sporą szanse na założenie konta
make_categorical_heatmap("job", "housing")
```

```python
make_categorical_heatmap("month", "education")
```
